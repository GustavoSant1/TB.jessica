##  Importações e Carregamento dos Dados

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker
import warnings
warnings.filterwarnings('ignore')

df = pd.read_csv('datatran2024.csv', sep=';', encoding='latin1')

print(f'✅ Dados carregados com sucesso! {df.shape[0]:,} linhas e {df.shape[1]} colunas.')
```

---
##  Tarefa Inicial – Verificação e Limpeza dos Dados

```python

print('Primeiras linhas:')
df.head()
```

```python

print('Informações gerais:')
df.info()
```

```python

ausentes = df.isnull().sum()
print(' Colunas com valores ausentes:')
ausentes[ausentes > 0]
```

```python

print(' Estatísticas descritivas:')
df.describe()
```

```python


df.columns = df.columns.str.strip().str.lower()


for col in ['veiculos', 'feridos_leves']:
    if col in df.columns:
        df[col] = pd.to_numeric(df[col].astype(str).str.replace(',', '.'), errors='coerce')

# Remove veículos com valor 0 ou negativo
antes = len(df)
df = df[df['veiculos'] > 0]
print(f' Removidas {antes - len(df)} linhas com veículos <= 0')

# Padroniza texto
df['dia_semana'] = df['dia_semana'].str.strip().str.capitalize()
df['uf'] = df['uf'].str.strip().str.upper()

print(f'Limpeza concluída! Dataset final: {len(df):,} registros.')
```

### Q1 – Média, Mediana e Moda do número de veículos envolvidos

```python
media   = df['veiculos'].mean()
mediana = df['veiculos'].median()
moda    = df['veiculos'].mode()[0]

print(f' Média  : {media:.2f} veículos')
print(f' Mediana: {mediana:.0f} veículo(s)')
print(f' Moda   : {moda:.0f} veículo(s) (valor mais frequente)')
```

```python
fig, ax = plt.subplots(figsize=(9, 4))

ax.hist(df['veiculos'].dropna(), bins=range(1, 12), color='#2E75B6', edgecolor='white', align='left')
ax.axvline(media,   color='#C00000', lw=2, linestyle='--', label=f'Média ({media:.1f})')
ax.axvline(mediana, color='#00B050', lw=2, linestyle=':',  label=f'Mediana ({mediana:.0f})')
ax.axvline(moda,    color='#FF8C00', lw=2, linestyle='-.', label=f'Moda ({moda:.0f})')

ax.set_title('Q1 – Distribuição de Veículos Envolvidos por Acidente', fontsize=13)
ax.set_xlabel('Número de veículos')
ax.set_ylabel('Frequência')
ax.legend()
ax.set_xticks(range(1, 11))
plt.tight_layout()
plt.savefig('q1_veiculos.png', dpi=150)
plt.show()
```

### Q2 – Desvio Padrão do número de feridos leves

```python
media_fl  = df['feridos_leves'].mean()
desvio_fl = df['feridos_leves'].std()
cv_fl     = (desvio_fl / media_fl) * 100

nivel = 'alta' if cv_fl > 30 else 'moderada' if cv_fl > 15 else 'baixa'

print(f' Média         : {media_fl:.2f} feridos leves')
print(f' Desvio Padrão : {desvio_fl:.2f}')
print(f' CV            : {cv_fl:.1f}%')
print(f' Interpretação : dispersão {nivel} em relação à média')
```

### Q3 – Tipo de acidente mais comum

```python
contagem_tipo = df['tipo_acidente'].value_counts()
mais_comum    = contagem_tipo.index[0]
pct           = contagem_tipo.iloc[0] / len(df) * 100

print(f' Tipo mais comum: {mais_comum}')
print(f' Representa {pct:.1f}% de todos os acidentes\n')
print('Top 5 tipos de acidente:')
contagem_tipo.head(5)
```

```python
top5 = contagem_tipo.head(5)

fig, ax = plt.subplots(figsize=(9, 4))
cores = ['#1F4E79'] + ['#2E75B6'] * 4
bars = ax.barh(top5.index[::-1], top5.values[::-1], color=cores[::-1], edgecolor='white')

for bar in bars:
    w = bar.get_width()
    ax.text(w + 50, bar.get_y() + bar.get_height()/2,
            f'{w:,.0f}', va='center', fontsize=9)

ax.set_title('Q3 – Top 5 Tipos de Acidente', fontsize=13)
ax.set_xlabel('Quantidade')
plt.tight_layout()
plt.savefig('q3_tipos.png', dpi=150)
plt.show()
```

### Q4 – Dia da semana com mais registros de acidentes

```python
ordem_dias = ['Segunda-Feira','Terça-Feira','Quarta-Feira',
              'Quinta-Feira','Sexta-Feira','Sábado','Domingo']

contagem_dia = df['dia_semana'].value_counts()
dias_ord = [d for d in ordem_dias if d in contagem_dia.index]
contagem_dia = contagem_dia.reindex(dias_ord).dropna()

dia_top = contagem_dia.idxmax()
print(f' Dia com mais acidentes: {dia_top} ({contagem_dia.max():,} registros)\n')
contagem_dia
```

```python
fig, ax = plt.subplots(figsize=(9, 4))
cores = ['#1F4E79' if d == dia_top else '#2E75B6' for d in contagem_dia.index]
bars = ax.bar(contagem_dia.index, contagem_dia.values, color=cores, edgecolor='white')

for bar in bars:
    h = bar.get_height()
    ax.text(bar.get_x() + bar.get_width()/2, h + 20,
            f'{h:,.0f}', ha='center', fontsize=8.5)

ax.set_title('Q4 – Acidentes por Dia da Semana', fontsize=13)
ax.set_ylabel('Nº de Acidentes')
ax.set_xticklabels(contagem_dia.index, rotation=30, ha='right')
plt.tight_layout()
plt.savefig('q4_dias.png', dpi=150)
plt.show()
```

### Q5 – Quantidade de acidentes por UF (gráfico de barras)

```python
acidentes_uf = df['uf'].value_counts().sort_values(ascending=False)

print(f'📌 UF com mais acidentes : {acidentes_uf.index[0]} ({acidentes_uf.iloc[0]:,})')
print(f'📌 UF com menos acidentes: {acidentes_uf.index[-1]} ({acidentes_uf.iloc[-1]:,})')
```

```python
fig, ax = plt.subplots(figsize=(14, 5))
cores = ['#1F4E79' if i == 0 else '#2E75B6' for i in range(len(acidentes_uf))]
bars = ax.bar(acidentes_uf.index, acidentes_uf.values, color=cores, edgecolor='white')

for bar in bars:
    h = bar.get_height()
    ax.text(bar.get_x() + bar.get_width()/2, h + 10,
            f'{h:,}', ha='center', va='bottom', fontsize=7)

ax.set_title('Q5 – Quantidade de Acidentes por UF', fontsize=14)
ax.set_xlabel('UF')
ax.set_ylabel('Nº de Acidentes')
ax.yaxis.set_major_formatter(mticker.FuncFormatter(lambda x, _: f'{int(x):,}'))
ax.set_xticklabels(acidentes_uf.index, rotation=45, ha='right')
plt.tight_layout()
plt.savefig('q5_acidentes_uf.png', dpi=150)
plt.show()
```
